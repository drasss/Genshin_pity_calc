### Un peu de maths

ici on a une loi multinomiale qui, a une occurence donnée (n= 76ème voeu par exemple) va donner une bernouilli de paramètre $p_n$
nous ici ce qu'on a ce sont les 90 $p_n$ associés

ce qu'il faut donc prendre en compte pour faire une loi, c'est la probabilité e d'arriver au point n (d'avoir fait n-1 echecs donc), multiplié par la probabilité $p_n$ d'avoir le personnage (multiplié par 1/2 pour le 50/50)

```python
e=[1]
r=[0]
```

```python

for i in range(len(pb)):
    r+=[e[-1]*pb[i]]
    e+=[e[-1]*(1-pb[i])]

```

```python

n=1
n5=1
result5=np.zeros((90))
result=np.zeros((180))
essai=500000
lost=0
np.random.seed(int(time.time()))
ct=0
while ct<essai:
    random=np.random.random()
    if random>pb[n5]:
        n+=1
        n5+=1
    else:
        ct+=1
        fifty=np.random.random()>=0.50
        result5[n5]+=1
        if fifty or lost:
            result[n]+=1
            n=1
        lost=1-lost
        n5=1
print(" resultat total : \n",result)
plt.plot(result)
plt.plot(result5)
result5
        

        
```
